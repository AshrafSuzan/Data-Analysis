```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Ashraf\Desktop\Jupyter\Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1,inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#use fillna() fuction to revmove NaN or NA from Amount Column
df['Amount'] = df['Amount'].fillna(0)
```

```python
df.head(10)
```

```python
#use drop() function to drop null column Status and unnamed
df.drop(['Status','unnamed1'], axis=1,inplace = True)
```

```python
df.head(10)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#use describe() function
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y= 'Amount',data= sales_gen)
```

```python
ax = sns.countplot (data= df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y= 'Amount',data= sales_Age)
```

```python
sales_State = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y= 'Orders',data= sales_State)
```

```python
sales_Amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y= 'Amount',data= sales_Amount)
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status', y= 'Amount',data= sales_status, hue='Gender')
```

```python
ax = sns.countplot(x = 'Occupation', data = df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Product_Category', data = df)

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_Category', y= 'Amount',data= sales_cat)
```
